```python
# Import dependencies
import os
import pandas as pd
import numpy as np
```

```python
# Read in csv
cdc_2015_df = pd.read_csv('2015.csv')
cdc_2015_df.head()
```

```python
# Drop columns that are entirely null
cdc_2015_df = cdc_2015_df.dropna(axis='columns', how='all')
cdc_2015_df.shape
```

```python
# Find columns with more than 50% null values
check = cdc_2015_df.isnull().sum() / len(cdc_2015_df) 

mostly_null_cols = check[check > 0.5].index

len(list(mostly_null_cols))
```
