```python
# Cleanup
clean_brfss_2015_df = brfss_2015_df[[
                                   'MENTHLTH', # Now thinking about your mental health, which includes stress, depression, and problems with emotions, for how many days during the past 30 days was your mental health not good?
                                   '_MICHD', # Respondents that have ever reported having coronary heart disease (CHD) or myocardial infarction (MI)
                                   '_RFHYPE5', # Adults who have been told they have high blood pressure by a doctor, nurse, or other health professional
                                   'TOLDHI2', # Have you EVER been told by a doctor, nurse or other health professional that your blood cholesterol is high?
                                   '_BMI5', # Body Mass Index (BMI)
                                   'SMOKE100', # Have you smoked at least 100 cigarettes in your entire life?
                                   'CVDSTRK3', # (Ever told) you had a stroke.
                                   'GENHLTH', # Would you say that in general your health is:
                                   'PHYSHLTH', # Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30 days was your physical health not good?
                                   'SEX', # Indicate sex of respondent
                                   '_AGEG5YR', # Fourteen-level age category
                                   ]]

print(clean_brfss_2015_df.shape)
clean_brfss_2015_df.head()
```

```python
# Drop null rows
clean_brfss_2015_df = clean_brfss_2015_df.dropna()
clean_brfss_2015_df.shape
```

```python
# Cleanup values using codebook: https://www.cdc.gov/brfss/annual_data/2015/pdf/codebook15_llcp.pdf
pd.options.mode.chained_assignment = None  # default='warn'

# MENTHLTH column
# change value 88 ("none") to 0
# remove 77 ("Donâ€™t know/Not sure")
# remove 99 ("Refused")
clean_brfss_2015_df['MENTHLTH'] = clean_brfss_2015_df['MENTHLTH'].replace({88:0})
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.MENTHLTH != 77]
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.MENTHLTH != 99]

# _MICHD column
# change 2 ("Did not report") to 0
clean_brfss_2015_df['_MICHD'] = clean_brfss_2015_df['_MICHD'].replace({2: 0})

# _RFHYPE5 column
# change 1 ("No") to 0
# change 2 ("Yes") to 1
# remove 9 ("Don't know/Not Sure/Refused/Missing")
clean_brfss_2015_df['_RFHYPE5'] = clean_brfss_2015_df['_RFHYPE5'].replace({1:0})
clean_brfss_2015_df['_RFHYPE5'] = clean_brfss_2015_df['_RFHYPE5'].replace({2:1})
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df._RFHYPE5 != 9]

# TOLDHI2 column
# change 2 ("No") to 0
# remove 7 ("Donâ€™t know/Not sure")
# remove 9 ("Refused")
clean_brfss_2015_df['TOLDHI2'] = clean_brfss_2015_df['TOLDHI2'].replace({2:0})
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.TOLDHI2 != 7]
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.TOLDHI2 != 9]

# SMOKE100 column
# change 2 ("No") to 0
# remove 7 ("Donâ€™t know/Not Sure")
# remove 9 ("Refused")
clean_brfss_2015_df['SMOKE100'] = clean_brfss_2015_df['SMOKE100'].replace({2:0})
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.SMOKE100 != 7]
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.SMOKE100 != 9]

# CVDSTRK3 column
# change 2 ("No") to 0
# remove 7 ("Donâ€™t know/Not Sure")
# remove 9 ("Refused")
clean_brfss_2015_df['CVDSTRK3'] = clean_brfss_2015_df['CVDSTRK3'].replace({2:0})
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.CVDSTRK3 != 7]
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.CVDSTRK3 != 9]

# GENHLTH column
# remove 7 ("Donâ€™t know/Not Sure")
# remove 9 ("Refused")
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.GENHLTH != 7]
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.GENHLTH != 9]

# PHYSHLTH column
# change 88 ("None") to 0
# remove 77 ("Donâ€™t know/Not Sure")
# remove 99 ("Refused")
clean_brfss_2015_df['PHYSHLTH'] = clean_brfss_2015_df['PHYSHLTH'].replace({88:0})
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.PHYSHLTH != 77]
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df.PHYSHLTH != 99]

# SEX column
# change 2 ("Female") to 0, because "Male" is already 1
clean_brfss_2015_df['SEX'] = clean_brfss_2015_df['SEX'].replace({2:0})

# _AGEG5YR column
# remove 14 ("Don't know/Refused/Missing")
clean_brfss_2015_df = clean_brfss_2015_df[clean_brfss_2015_df._AGEG5YR != 14]

clean_brfss_2015_df.head()
```

```python
# Cleanup column names
clean_brfss_2015_df = clean_brfss_2015_df.rename(columns = {
                                                        'MENTHLTH':'MentalHealthDays', 
                                                        '_MICHD':'ReportedHeartDiseaseAttack', 
                                                        '_RFHYPE5':'HighBP',  
                                                        'TOLDHI2':'HighChol', 
                                                        '_BMI5':'BMI', 
                                                        'SMOKE100':'Smoker', 
                                                        'CVDSTRK3':'Stroke', 
                                                        'GENHLTH':'GeneralHealth', 
                                                        'PHYSHLTH':'PhysicalHealth', 
                                                        'SEX':'Sex', 
                                                        '_AGEG5YR':'Age', 
                                                        })

clean_brfss_2015_df.head()
```
